```python
news = news[['date', 'title','text']]
news.head()
```

```python
#Remova NaN
news=news.dropna()
news.head()
```

```python
#dates format
def convert_date(date):
    try:
        return pd.to_datetime(date, format='%d %B %Y').strftime('%Y-%m-%d')
    except ValueError:
        try:
            return pd.to_datetime(date, format='%d/%m/%Y').strftime('%Y-%m-%d')
        except ValueError:
            greek_months = {
                'Ιανουαρίου': 'January',
                'Φεβρουαρίου': 'February',
                'Μαρτίου': 'March',
                'Απριλίου': 'April',
                'Μαΐου': 'May',
                'Ιουνίου': 'June',
                'Ιουλίου': 'July',
                'Αυγούστου': 'August',
                'Σεπτεμβρίου': 'September',
                'Οκτωβρίου': 'October',
                'Νοεμβρίου': 'November',
                'Δεκεμβρίου': 'December'
            }
            for greek_month, english_month in greek_months.items():
                if greek_month in date:
                    date = date.replace(greek_month, english_month)
                    break
            return pd.to_datetime(date, format='%d %B %Y').strftime('%Y-%m-%d')
```

```python
# Convert the dates to the desired format
news['date'] = news['date'].apply(lambda x: convert_date(x))

# Print the updated DataFrame
print(news)
```

```python
# Filter the headlines for the year 2022
news_2022 = news[news['date'].str.startswith('2022')]
news_2022.head()
```

```python
len(news_2022)
```

```python
#Convert the df to list of articles
docs=news_2022.text.tolist()
docs[0]
```

```python
from top2vec import Top2Vec
```

```python
model=Top2Vec(docs)
#Top2Vec expects a list of different strings as input, that's why we converted the df to list, and each string is a unique document
```

```python
topic_sizes,topic_nums=model.get_topic_sizes()
print(topic_sizes)
print()
print(topic_nums)
```

```python
topic_words, word_scores, topic_nums=model.get_topics(10) #get the top 10 topics
```

```python
for words, scores,num in zip(topic_words, word_scores,topic_nums):
  print(num)
  print(f"Words: {words}")
```

```python
#Start looking at specific topics and at the documents that correspond to it
documents, document_scores, document_ids=model.search_documents_by_topic(topic_num=0, num_docs=10)
```

```python
for doc,score,doc_id in zip(documents, document_scores,document_ids):
  print(f"Document: {doc_id}, Score:{score}")
  print("________")
  print(doc)
  print("_________")
  print()
```
